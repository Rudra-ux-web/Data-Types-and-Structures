Q1: What are data structures, and why are they important?

A: Data structures are ways to organize and store data so that it can be used efficiently. They are important because they make operations like searching, sorting, inserting, and deleting data faster and easier.

Q2: Explain the difference between mutable and immutable data types with examples.

A: Mutable data types can be changed after creation (e.g., list, dictionary, set). Immutable data types cannot be changed after creation (e.g., string, tuple, int).

Mutable
lst = [1, 2, 3]
lst.append(4)
print(lst)  # [1, 2, 3, 4]
Immutable
s = "hello"
s = s.upper()
print(s)  # HELLO


Q3: What are the main differences between lists and tuples in Python?

List: Mutable, can change elements, written as [].

Tuple: Immutable, cannot change elements, written as ().

Q4: Describe how dictionaries store data.

A: Dictionaries store data as key-value pairs. Each key is unique and maps to a value. Internally, they use hash tables for fast access.

Q5: Why might you use a set instead of a list in Python?

A: A set is used when we need to store unique elements without duplicates and perform mathematical operations like union or intersection.

Q6: What is a string in Python, and how is it different from a list?

A: A string is a sequence of characters enclosed in quotes. It is immutable. A list is a sequence of items and is mutable.

Q7: How do tuples ensure data integrity in Python?

A: Since tuples are immutable, the stored data cannot be accidentally changed, which ensures data integrity.

Q8: What is a hash table, and how does it relate to dictionaries in Python?

A: A hash table stores data using a hash function that maps keys to values. Python dictionaries are implemented using hash tables for fast access.

Q9: Can lists contain different data types in Python?

A: Yes, lists can hold mixed data types, e.g., [1, "apple", 3.14].

Q10: Explain why strings are immutable in Python.

A: Strings are immutable because once created, they cannot be changed. Any modification creates a new string.

Q11: What advantages do dictionaries offer over lists for certain tasks?

A: Dictionaries provide faster lookups using keys, while lists require searching through elements.

Q12: Describe a scenario where using a tuple would be preferable over a list.

A: When we want to store fixed data that should not change, like coordinates (x, y, z) or days of the week.

Q13: How do sets handle duplicate values in Python?

A: Sets automatically remove duplicates. Example: {1,2,2,3} becomes {1,2,3}.

Q14: How does the “in” keyword work differently for lists and dictionaries?

A: For lists, in checks if a value exists. For dictionaries, in checks if a key exists.

Q15: Can you modify the elements of a tuple? Explain why or why not.

A: No, you cannot modify a tuple because it is immutable.

Q16: What is a nested dictionary, and give an example of its use case.

A: A dictionary inside another dictionary.

student = {
    "name": "Ravi",
    "marks": {"math": 90, "science": 85}
}


Q17: Describe the time complexity of accessing elements in a dictionary.

A: Accessing an element by key is O(1) on average because of hashing.

Q18: In what situations are lists preferred over dictionaries?

A: Lists are preferred when order matters or when we don’t need key-value mapping.

Q19: Why are dictionaries considered unordered, and how does that affect data retrieval?

A: Dictionaries retrieve data by keys, not positions. Order preservation only came in Python 3.7+.
